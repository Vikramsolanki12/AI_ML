```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
pivot_table = tips.pivot_table(
    values ="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)
```

```python
pivot_table
```

```python
sns.heatmap(
    pivot_table,
    cmap="coolwarm",
    annot=True,
    fmt=".2f"
)
plt.title("day time combination")
plt.show()
```
