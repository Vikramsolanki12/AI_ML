```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
data = sns.load_dataset("tips")
```

```python
fig , ax = plt.subplots()
sns.scatterplot(
    data=data,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("total_bill vs tip")
plt.show()
```
